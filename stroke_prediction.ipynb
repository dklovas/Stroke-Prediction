11. smoking_status: Patient’s smoking habits (Categorical: "formerly smoked", "never smoked", "smokes", "Unknown").

12. stroke: Target variable indicating whether the patient has had a stroke (Binary: 1 = Yes, 0 = No).

Key Insights from the Dataset:

## 3.5. Variance Inflation Factor (VIF)

Interpretation of VIF:
- VIF = 1: No correlation between the feature and other features.
- 1 < VIF < 5: Moderate correlation, but generally acceptable.
- VIF > 5: High correlation, might cause multicollinearity problems.
- VIF > 10: Strong multicollinearity, which could affect the model performance, and it's often recommended to remove or transform such features.

```python
def get_vif(df):
    X = df[df.select_dtypes(exclude=["object"]).columns.to_list()]
    X = sm.add_constant(X)

    vif_data = pd.DataFrame(
        {
            "Feature": X.columns,
            "VIF": [variance_inflation_factor(X.values, i) for i in range(X.shape[1])],
        }
    )

    return vif_data


get_vif(df)
```

- no multicollinearity in dataset

## 3.6. Hypotesis Results

#### T-Test

#### Assumption:
- Independence: Satisfied as user IDs are unique with no overlap between groups.  
- Normality: Distribution have to be normal, but due to a large sample size supportes the Central Limit Theorem.
- Equal Variances: Assessed with Levene’s test. If p-value > 0.05, equal variances assumed; otherwise, use Welch's t-test (`equal_var=False`). 

#### T-Test for `age` vs `stroke`

```python
feature = "age"
target = y_col

data_1 = df[df[target] == 1][feature]
data_2 = df[df[target] == 0][feature]

stat, p_value_levene = levene(data_1, data_2)
print(f"Test Statistic (Levene’s)={stat:.4f}, p-value={p_value_levene:.4f} \n")

results = two_sides_ttest_with_ci(data_1, data_2, equal_var=False)
print_two_sides_ttest_with_ci(results)
```

Hypothesis Test Results: Stroke and Age:
- The result of Levene's test indicates that the variances between the two groups (stroke = 0 and stroke = 1) are significantly different, as the p-value is very small (p < 0.05). So have to use Welch's t-test (`equal_var=False`).
- The T-test results show a significant difference in the mean age between individuals with and without a stroke, as the p-value is very small (p < 0.05). The mean age for individuals with a stroke is significantly higher than for those without. The confidence interval for the mean difference (24.05, 27.463) does not contain zero, further confirming the existence of a significant difference.


#### T-Test for `bmi` vs `stroke`

```python
feature = "bmi"
target = y_col

data_1 = df[df[target] == 1][feature]
data_2 = df[df[target] == 0][feature]

stat, p_value_levene = levene(data_1, data_2)
print(f"Test Statistic (Levene’s)={stat:.4f}, p-value={p_value_levene:.4f} \n")

results = two_sides_ttest_with_ci(data_1, data_2, equal_var=False)
print_two_sides_ttest_with_ci(results)
```

Hypothesis Test Results: BMI vs Stroke:

- The result of Levene's test indicates that the variances between the two groups (stroke = 0 and stroke = 1) are significantly different, as the p-value is very small (p < 0.05). Therefore, we reject the null hypothesis of equal variances.
- The T-test results show a significant difference in the mean BMI between individuals with and without a stroke, as the p-value is very small (p < 0.05). The mean BMI for individuals with a stroke is significantly higher than for those without. The confidence interval for the mean difference (0.642, 2.198) does not contain zero, further confirming the existence of a significant difference.

## 5.2. Model selection

### Model Pipelines, Preprocessing, and Cross-Validation

Used:
1. Class Weight Balancing:
   - class_weight='balanced' is used to handle imbalanced datasets by adjusting weights inversely proportional to class frequencies.

2. Feature Preprocessing:
   - All features in X_train are scaled using StandardScaler within a ColumnTransformer. Remaining columns (if any) are passed through unchanged.

3. Model Dictionary:
   - A dictionary, models_with_scaling, defines multiple classifiers, including:
     - LogisticRegression
     - SVC
     - KNeighborsClassifier
     - HistGradientBoostingClassifier
     - LGBMClassifier
     - GradientBoostingClassifier
     - BaggingClassifier (with DecisionTreeClassifier as the base estimator).

4. Pipeline Creation:
   - The create_pipeline_with_scaling function constructs a pipeline for each model:
     - Includes the preprocessor_with_scaling.
     - Appends the classifier as the final step.


```python
class_weight = "balanced"

features_for_StandardScaler = X_train.columns

preprocessor_with_scaling = ColumnTransformer(
    transformers=[("standard", StandardScaler(), features_for_StandardScaler)],
    remainder="passthrough",
)

models_with_scaling = {
    "LogisticRegression": LogisticRegression(class_weight=class_weight),
    "SVC": SVC(class_weight=class_weight),
    "KNN": KNeighborsClassifier(),
    "HistGradientBoosting": HistGradientBoostingClassifier(class_weight=class_weight),
    "LGBMClassifier": lgb.LGBMClassifier(class_weight=class_weight, verbose=-1),
    "GradientBoostingClassifier": GradientBoostingClassifier(),
    "BaggingClassifier": BaggingClassifier(
        estimator=DecisionTreeClassifier(class_weight=class_weight), n_estimators=10
    ),
}


def create_pipeline_with_scaling(model):
    return Pipeline(
        steps=[("preprocessor", preprocessor_with_scaling), ("classifier", model)],
        memory="./cache_directory",
    )


kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
accuracy = {"model": [], "recall class 1": []}
results_model = {}
confusion_matrix_model_selection = {}
confusion_matrix_tuned = {}
model_reports = []

for model_name, model in models_with_scaling.items():

    pipeline = create_pipeline_with_scaling(model)
    scores = cross_validate(pipeline, X_train, y_train, cv=kf, scoring=scoring)
    results_model[model_name] = get_data_from_cross_validate(scores)

    accuracy["model"].extend([model_name] * len(scores["test_accuracy"]))
    accuracy["recall class 1"].extend(scores["test_recall"])

    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_val)
    report = classification_report(y_val, y_pred, output_dict=True, zero_division=0)
    model_reports.append(get_data_from_classification_report(report, 1, model_name))

    confusion_matrix_model_selection[model_name] = confusion_matrix(y_val, y_pred)
    confusion_matrix_tuned[model_name] = np.array([[0, 0], [0, 0]])

results_init = pd.DataFrame.from_dict(results_model, orient="index")
results_init = results_init.reset_index().rename(columns={"index": "Model"})

model_reports_df = pd.DataFrame(model_reports)
display(model_reports_df)
```

### LogisticRegression

```python
best_estimators = {}

kf = KFold(n_splits=5, shuffle=True, random_state=42)

Logistic_Regression_grid = {
    "classifier__C": uniform(0.01, 10),
    "classifier__penalty": ["l2"],
    "classifier__solver": ["lbfgs", "liblinear", "saga"],
    "classifier__max_iter": [200, 500],
}

model = LogisticRegression(class_weight=class_weight)
model_name = model.__class__.__name__

pipeline = create_pipeline_with_scaling(model)

search = RandomizedSearchCV(
    estimator=pipeline,
    param_distributions=Logistic_Regression_grid,
    n_iter=100,
    scoring="accuracy",
    cv=kf,
    verbose=1,
    random_state=42,
    n_jobs=-1,
)

search.fit(pd.concat([X_train, X_val]), pd.concat([y_train, y_val]))
best_estimators[model_name] = search.best_estimator_
print(f"\nBest params for {model_name}:")
for param, value in search.best_params_.items():
    print(f"  - {param.split('__')[-1]}: {value}")
print(f"\nAccuracy for {model_name}: {search.best_score_:.3f}")
```

After performing grid search, the following hyperparameters were identified as optimal for the LogisticRegression model:

1. C:
   - The regularization strength. A smaller value of C corresponds to stronger regularization, which helps prevent overfitting by discouraging large coefficients.
   - In this case, C is close to 1, indicating moderate regularization, balancing bias and variance.

2. max_iter:
   - The maximum number of iterations the algorithm will perform before stopping.
   - This parameter ensures the solver has enough iterations to converge, particularly for complex datasets. Setting this to 500 allows the model to fully optimize without prematurely halting.

3. penalty:
   - The type of regularization applied. 
   - l2 regularization (Ridge penalty) adds a penalty term proportional to the squared magnitude of coefficients, helping control overfitting and ensuring stability.

4. solver: 
   - The optimization algorithm used to minimize the loss function. 
   - saga is a fast, stochastic solver that works well with l2 regularization and is suitable for large datasets or datasets with sparse features.

#### Model Performance
- Accuracy:
   - Indicates that the model correctly classifies approximately 73% of the validation samples.

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)

SVC_grid = {
    "classifier__C": uniform(0.1, 10),
    "classifier__kernel": ["linear", "rbf", "poly", "sigmoid"],
    "classifier__gamma": ["scale", "auto"],
    "classifier__degree": randint(2, 5),
    "classifier__coef0": uniform(0, 1),
    "classifier__shrinking": [True, False],
    "classifier__tol": uniform(1e-5, 1e-3),
}

model = SVC(probability=True, class_weight=class_weight)
model_name = model.__class__.__name__

pipeline = create_pipeline_with_scaling(model)

search = RandomizedSearchCV(
    estimator=pipeline,
    param_distributions=SVC_grid,
    n_iter=100,
    scoring="accuracy",
    cv=kf,
    verbose=1,
    random_state=42,
    n_jobs=-1,
)

search.fit(pd.concat([X_train, X_val]), pd.concat([y_train, y_val]))
best_estimators[model_name] = search.best_estimator_

print(f"\nBest params for {model_name}:")
for param, value in search.best_params_.items():
    print(f"  - {param.split('__')[-1]}: {value}")
print(f"\nAccuracy for {model_name}: {search.best_score_:.3f}")
```

Insights on SVC Model After Hyperparameter Tuning

Best Hyperparameters for SVC:
- C:
    - Regularization parameter. A high value indicates a lower regularization strength, allowing the model to fit more closely to the data. In this case, it helps improve the model's ability to classify even challenging cases, especially in the context of a non-linear kernel.
- kernel:
    - The polynomial kernel allows the SVC to model non-linear relationships, which is especially useful when the data isn't linearly separable.
- degree:
    - The degree of the polynomial kernel. A degree of 4 suggests that the decision boundary is quite flexible, making the model more powerful in learning complex patterns in the data.
- coef0:
    - The coefficient that influences the kernel function. For a polynomial kernel, this controls the impact of higher-degree terms. A value of 0.34 helps in controlling the flexibility of the decision boundary.
- gamma:
    - This parameter determines how much influence a single training example has. scale automatically calculates gamma as 1 / (n_features * X.var()), allowing the model to learn in a balanced way based on feature variance.
- shrinking: False
    - Whether to use the shrinking heuristic to speed up training. Setting this to False means the model is trained without the heuristic.
- tol:
    - The tolerance for stopping criteria. This small value ensures the model trains until a very accurate decision boundary is found, as long as improvements are significant enough.

Performance After Tuning:

5. learning_rate:

6. colsample_bytree:

## 5.5. Optimizing LogisticRegression Model

### ROC Curve and AUC for Logistic Regression

- The Receiver Operating Characteristic (ROC) curve is a graphical representation of a classifier's performance. It plots the True Positive Rate (TPR) against the False Positive Rate (FPR) at various threshold values.


```python
model = best_estimators["LogisticRegression"]

y_prob = model.predict_proba(X_test)[:, 1]

fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

plot_roc(roc_auc, fpr, tpr)
```

- With an AUC of 0.88 the model has proven its effectiveness in classifying the target variable correctly, making it a reliable choice for detecting strokes in the dataset.

### Evaluation of Optimal Threshold for Logistic Regression Model

```python
confusion_matrix_Optimal = {}

youden_j = tpr - fpr
optimal_threshold = thresholds[np.argmax(youden_j)]

print(f"Optimal Threshold (maximizing TPR - FPR): {optimal_threshold:.4f}")

y_pred_default = (y_prob >= 0.5).astype(int)
y_pred_optimal = (y_prob >= optimal_threshold).astype(int)

conf_matrix_default = confusion_matrix(y_test, y_pred_default)
conf_matrix_optimal = confusion_matrix(y_test, y_pred_optimal)

accuracy_default = accuracy_score(y_test, y_pred_default)
precision_default = precision_score(y_test, y_pred_default)
recall_default = recall_score(y_test, y_pred_default)

accuracy_optimal = accuracy_score(y_test, y_pred_optimal)
precision_optimal = precision_score(y_test, y_pred_optimal)
recall_optimal = recall_score(y_test, y_pred_optimal)

print("\nPerformance Metrics (Default Threshold - 0.5):")
print(f"Accuracy: {accuracy_default:.4f}")
print(f"Precision: {precision_default:.4f}")
print(f"Recall: {recall_default:.4f}")

print(f"\nPerformance Metrics (Optimal Threshold): {optimal_threshold:.4f}")
print(f"Accuracy: {accuracy_optimal:.4f}")
print(f"Precision: {precision_optimal:.4f}")
print(f"Recall: {recall_optimal:.4f}")

confusion_matrix_Optimal["Default Threshold - 0.5"] = conf_matrix_default
confusion_matrix_Optimal[f"Optimal Threshold - {optimal_threshold:.4f}"] = (
    conf_matrix_optimal
)
plot_confusion_matrix_heatmaps(confusion_matrix_Optimal, 1, 2, subplot_size=(8, 3))
```

The optimal threshold that maximizes TPR - FPR was found to be 0.63.

Insights:
- Recall is maximized at the default threshold, indicating the model is very good at detecting stroke cases, but at the expense of precision. The high recall means that very few stroke cases are missed.
- With the optimal threshold, the balance between recall and precision improves slightly, with the model achieving a better overall accuracy (0.8043) while still maintaining a good recall of 0.8800.
- The confusion matrix shows that with the optimal threshold, there are fewer false positives, resulting in better classification of the non-stroke class, though some stroke cases are missed.
   
Conclusion:
- If the priority is minimizing false negatives (i.e., detecting as many stroke cases as possible), the default threshold is better because it maximizes recall.
- However, if a more balanced performance is desired, with higher accuracy and better precision, the optimal threshold (0.6278) offers a more refined model.
